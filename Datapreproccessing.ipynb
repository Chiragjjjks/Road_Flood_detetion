```python
import pandas as pd 
df =pd.read_csv("Dataset/bangalore_urban_flood_prediction_AI.csv")
df
```

```python
from sklearn.preprocessing import MinMaxScaler

# Select features to scale (excluding Latitude, Longitude)
features_to_scale = ['Altitude', 'Rainfall_Intensity', 'Temperature', 'Humidity', 
                     'Atmospheric_Pressure', 'River_Level', 'Drainage_Capacity', 
                     'Population_Density', 'Urbanization_Level']

scaler = MinMaxScaler()
df[features_to_scale] = scaler.fit_transform(df[features_to_scale])

# Keep Latitude and Longitude unchanged
X = df.drop(columns=['flood'])  # Features
y = df['flood']  # Target variable
```
