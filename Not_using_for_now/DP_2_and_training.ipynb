```python
import pandas as pd
import numpy as np

file_path = "C:/Users/chira/Desktop/on_going_projects/Final_year/Dataset/bangalore_urban_flood_prediction_AI.csv"  
df = pd.read_csv(file_path)
df.head()
```

```python
X_real_time = df[['Latitude', 'Longitude', 'Temperature', 'Humidity', 'Atmospheric_Pressure']]
y = df['flood']
```

```python
X_long_lan = df[['Latitude','Longitude']]
X_long_lan
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_real_time, y, test_size=0.2, random_state=42)
X_test,X_train
```

```python
print(df.isnull().sum())  
df = df.dropna()
df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)

X_train_scaled
```

### Train MLP and XGBoost

## XGBoost training 

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score

xgb_model = xgb.XGBClassifier(n_estimators=100, max_depth=5, learning_rate=0.001)
xgb_model.fit(X_train, y_train)

# Predict
y_pred_xgb = xgb_model.predict(X_test)
print("XGBoost Accuracy:", accuracy_score(y_test, y_pred_xgb))
```

## MLP training

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

mlp_model = Sequential([
    Dense(100, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.2),
    Dense(100, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])

mlp_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
mlp_model.fit(X_train_scaled, y_train, epochs=50, batch_size=16, validation_data=(X_test_scaled, y_test))
```

```python
# Define early stopping
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=100, restore_best_weights=True)
```

```python
## Train the model
history=mlp_model.fit(X_train,y_train,epochs=100,validation_data=(X_test,y_test),verbose=1,callbacks=[early_stopping])
```
